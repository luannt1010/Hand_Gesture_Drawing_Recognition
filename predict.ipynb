```python
import tensorflow as tf
import numpy as np
from preprocess_image import preprocess
import os
import matplotlib.pyplot as plt
```

```python
classes = sorted(os.listdir("dataset"))
img_path = "test.png"
model_path = r"models/best_model3.h5"
model = tf.keras.models.load_model(model_path)
img, tensor = preprocess(img_path=img_path, is_predict=True)
preds = model.predict(tensor)
print(f"Class: {classes[np.argmax(preds[0])]} {np.max(preds[0])}")
plt.imshow(img)
```
